```python
import numpy as np
import os
import nltk
from nltk.corpus import stopwords
from tqdm.notebook import tqdm
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import random
```

```python
filename = 'DataQ4.txt'
f = open(filename, 'r')
text = f.read()
f.close()
```

```python
text_list = text.strip().split("\n")
text_list
```

```python
Y_test = []
for i in tqdm(range(len(text_list))):
    Y_test.append(int(text_list[i][-1]))
    text_list[i] = text_list[i][:len(text_list[i])-3] # remove the last three letters from each sentence
    text_list[i] = text_list[i].strip()
Y_test = np.array(Y_test)
Y_test
```

```python
#int(text_list[0][-1])
Y_test.shape
```

```python
text_list = list(np.char.lower(text_list))
text_list
```

```python
# formig the set of unique words and eliminating stopwords and words having a single letter 
words = set()

for review in text_list:
    for word in review.split(" "):
        if len(word) <= 1 and word in stopwords.words('english'):
            continue
        words.add(word)
                                  
print(len(words))
```

```python
# formimg term-frequency ; i.e. TF
flag = False
for review in text_list:
    temp = np.zeros((1,len(words)))  # stores the frequency of occurance of words from the set of unique words
    #index = 0
    for index,word in enumerate(words):
        cnt = 0
        for element in review.split(" "):
            cnt += int(word == element)
        temp[0,index] = cnt
        #index += 1
    if flag==False:
        TF=temp
        flag = True
        continue
    TF = np.concatenate((TF,temp),axis = 0)

        
#print(TF.shape)

Review_len = np.dot(TF,np.ones((len(words),1))) # stores the length of each review, shape = (1000,1)
#print(Review_len.shape)
```

```python
i = 0
for i in tqdm(range(len(TF))):
    TF[i,:] = TF[i,:] / Review_len[i,0]
    i += 1
#TF
```

```python
def K_means(n_classes,X):

    old_mew = np.zeros((n_classes,X.shape[1]))
    mew = []
    
    for i in range(n_classes):
        mew.append(X[random.randint(0,X.shape[0]),:])

    mew = np.array(mew)
    print("mew = ",mew.shape)

    D = np.zeros((X.shape[0],n_classes)) # stores info related to the class a point belongs to

    #print(D)
    while True:
        print("Distance matrix calculation....")
        for i in tqdm(range(X.shape[0])):
            dist = 1e12
            index = 0
            for j in range(n_classes):
                D[i,j] = 0
                if dist > np.dot(X[i,:] - mew[j], X[i,:] - mew[j]):
                    dist = np.dot(X[i,:] - mew[j], X[i,:] - mew[j])
                    index = j
            D[i,index] = 1
            
        #print(D)
        print("Update mew...")
        for i in tqdm(range(n_classes)):
            mew[i] = np.mean(X[D[:,i] == 1])#/np.sum(D[:,i])


        flag = False 
        for i in range(n_classes):
            if np.abs(np.dot(old_mew[i],old_mew[i]) - np.dot(mew[i],mew[i])) > 1e-3:
                flag = True
                break

        if flag == False:
#             plt.scatter(X[D[:,0]==1,0],X[D[:,0]==1,1],color='b')
#             plt.scatter(X[D[:,0]==0,0],X[D[:,0]==0,1],color='r')
#             plt.show()
            return mew

        old_mew = mew.copy()
```

```python
# initialising and calculating DF
DF = np.zeros((1,len(words)))
#DF.shape
i = 0
cnt = 0
for i in range(len(words)):
    cnt = 0
    for j in range(TF.shape[0]):
        cnt += int(TF[j,i] > 0)
    DF[0,i] = TF.shape[0]/(cnt + 1)
DF = np.log(DF)
DF.shape
```

```python
# forming TF-iDF matrix
print(TF.shape)
print(DF.shape)
TF_iDF = np.multiply(TF,DF)
TF_iDF.shape
```

```python
#PCA
n_components = 10
pca_text = PCA(n_components = n_components)
X = pca_text.fit_transform(TF_iDF)
X.shape
```

```python
# plot first two components of X
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
# initialisation before EM
#means = K_means(2,X)
#print(means.shape)
mew_1 = X[0,:]
mew_2 = X[5,:]
sigma_1 = np.dot((X - mew_1).T,X-mew_1)
sigma_2 = np.dot((X - mew_2).T,X-mew_2)
pi_1 = 0.5
pi_2 =  0.5


# posterior
gamma_1 = np.zeros((TF.shape[0],))
gamma_2 = np.zeros((TF.shape[0],))

#sigma_1.shape
```

```python
n_iter = 15
likelihood = []
for i in tqdm(range(n_iter)):
    print(f"Iteration number = {i}")
    print("E-step.....")
    for i in tqdm(range(TF.shape[0])):
        gamma_1[i] = (pi_1 * multivariate_normal.pdf(X[i,:], mean=mew_1, cov=sigma_1,allow_singular = True))/(pi_1*multivariate_normal.pdf(X[i,:], mean=mew_1, cov=sigma_1,allow_singular = True) + pi_2 * multivariate_normal.pdf(X[i,:], mean=mew_2, cov=sigma_2,allow_singular = True))
        gamma_2[i] = (pi_2 * multivariate_normal.pdf(X[i,:], mean=mew_2, cov=sigma_2,allow_singular = True))/(pi_1*multivariate_normal.pdf(X[i,:], mean=mew_1, cov=sigma_1,allow_singular = True) + pi_2 * multivariate_normal.pdf(X[i,:], mean=mew_2, cov=sigma_2,allow_singular = True))
        
        
    print("M step......")
    mew_1 = np.matmul(X.T,gamma_1)/np.sum(gamma_1)
    mew_2 = np.matmul(X.T,gamma_2)/np.sum(gamma_2)

    temp_mat1 = np.zeros((n_components,n_components))
    temp_mat2 = np.zeros((n_components,n_components))
    
    print("sigma")
    for i in tqdm(range(TF.shape[0])):
        x = (X[i,:] - mew_1).reshape(-1,1)  # column vector
        temp_mat1 += gamma_1[i]* (x @ x.T)
        y = (X[i,:] - mew_2).reshape(-1,1)  # column vector
        temp_mat2 += gamma_2[i] * ( y @ y.T)
    
    # diagonal covariance matrix computation
    sigma_1 = np.diag(np.diag(temp_mat1/np.sum(gamma_1)))  # diagonal covariance
    sigma_2 = np.diag(np.diag(temp_mat2/np.sum(gamma_2)))  # diagonal covariance
    
    # posteriors for two classes
    pi_1 = np.mean(gamma_1)
    pi_2 = np.mean(gamma_2) 
    
    # plot of first two dimensions at each iterations...
    plt.scatter(X[gamma_1 > gamma_2,0],X[gamma_1 > gamma_2,1],color= 'b')
    plt.scatter(X[gamma_1 <= gamma_2,0],X[gamma_1 <= gamma_2,1],color= 'r')
    plt.show()
    
    print("Likelihood evaluation")
    like = 0
    for i in tqdm(range(TF.shape[0])):
        like += np.log(1e-12 + pi_1 * multivariate_normal.pdf(X[i,:], mean=mew_1, cov=sigma_1,allow_singular = True)+ pi_2 * multivariate_normal.pdf(X[i,:], mean=mew_2, cov=sigma_2,allow_singular = True))
    print("Likelihood = ",like)                                                                  
    likelihood.append(like)
        
    
```
